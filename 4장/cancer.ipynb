```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 셋 로드- cancer_b.csv(양성데이터), cancer_m.csv(악성데이터)
cancer_b_data = pd.read_csv("https://raw.githubusercontent.com/zzhining/data_visualization/main/dataset/cancer_b.csv?token=GHSAT0AAAAAAB4ACKNOBJ5MCFOFJG2HLKS6Y4N2AKA", index_col="Id")
cancer_m_data = pd.read_csv("https://raw.githubusercontent.com/zzhining/data_visualization/main/dataset/cancer_m.csv?token=GHSAT0AAAAAAB4ACKNOGZ4ZJYBRMSNPMBX6Y4N2AVQ", index_col="Id")
cancer = pd.concat([cancer_b_data, cancer_m_data])
```

```python
sns.histplot(data=cancer, x = 'Area (mean)', hue='Diagnosis')
```

```python
sns.kdeplot(data=cancer, x = 'Area (mean)', hue='Diagnosis')
```

```python
sns.histplot(data=cancer, x = 'Area (mean)', hue='Diagnosis', bins = 100)
```

```python
sns.histplot(data=cancer, x = 'Area (mean)', kde = True)
```
